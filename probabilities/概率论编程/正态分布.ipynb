```python
import numpy
#numpy.random.normal(size=100)
import random
# 生成呈正态分布的随机数
# print("normalvariate: ", random.normalvariate(0, 1))
n=eval(input("请输入n："))
mu=eval(input("请输入mu："))
sigma=eval(input("请输入sigma："))
# 产生一组满足正态分布的随机数
group = []
for _ in range(n):
    group.append(random.normalvariate(mu, sigma))
# 画成直方图
import matplotlib.pyplot as plt  # 导入模块
plt.hist(group, bins=30)  # bins直方图的柱数
plt.show()
print(group)
```

```python
import scipy
from scipy import stats
scipy.stats.normaltest (group, axis=0)
#scipy.stats.anderson (group, dist ='norm' )
```
